```python
!pip install tf2onnx onnx

import os
# 1. The GPU Blindfold
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import tensorflow as tf
import tf2onnx
import onnx

# CHANGE THIS if your uploaded file is named differently
keras_model_path = "bilstm_model.h5"
onnx_output_path = "bilstm_model.onnx"

print("⏳ Loading model...")
model = tf.keras.models.load_model(keras_model_path)

# 2. THE BYPASS: Trace the raw math operations, ignoring Keras metadata completely
print("⏳ Tracing raw TensorFlow operations...")
@tf.function(input_signature=[tf.TensorSpec([None, 50], tf.int32, name="input_ids")])
def trace_model(inputs):
    return model(inputs)

# 3. Convert the pure math function, not the Keras object
print("⏳ Converting pure math graph to ONNX...")
onnx_model, _ = tf2onnx.convert.from_function(
    trace_model,
    input_signature=[tf.TensorSpec([None, 50], tf.int32, name="input_ids")],
    opset=13
)

onnx.save(onnx_model, onnx_output_path)
print("✅ Success! Check the file menu on the left.")
```
